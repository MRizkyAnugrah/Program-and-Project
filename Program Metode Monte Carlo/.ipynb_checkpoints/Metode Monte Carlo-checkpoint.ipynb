## Fungsi

```python
import numpy as np
import sympy as smp
import matplotlib.pyplot as plt

def f_v(t):
    vt = np.exp(-t)*np.cos(t)
    return vt

def trapezoidal(function_trap,a,b,h):
    if a > b:
        print('the value of b must be bigger than a') 
        return 'ErrorValue'
    N = int((b-a)/h)
    func_a = function_trap.subs({sym_t:a}).evalf()
    func_b = function_trap.subs({sym_t:b}).evalf()
    s = 0
    for i in range (1,N):
        s += function_trap.subs({sym_t:i}).evalf()
    IntegralTrap = (h/2)*(func_a +  2*s + func_b)
    div_2 = smp.diff(function_trap,sym_t,2)
    jumlah_div2 = 0
    steps_error = np.arange(a,b,h)
    for i in steps_error:
        jumlah_div2 += div_2.subs({sym_t:((i+(i+h))/2)}).evalf()
    error_Trap = (b-a)/12*h**2*jumlah_div2
    return IntegralTrap, error_Trap

def montecarlo(f,a,b,n,option="uniform"):
    np.random.seed(314159)
    us = np.random.uniform(a, b, n)
    
    if option=="uniform":
        fs = f(us)
    else:
        c0 = 4 - 2*np.sqrt(2)
        c1 = -6 + 4*np.sqrt(2)
        xs = (-c0 + np.sqrt(2*c1*us + c0**2))/c1
        fs = f(xs)/(c0 + c1*xs)
        
    fbar, err = stats(fs)
    return (b-a)*fbar, (b-a)*err

def stats(fs):
    n = fs.size
    fbar = np.sum(fs)/n
    fsq = np.sum(fs**2)/n
    varfbar = (fsq - fbar**2)/(n - 1)
    return fbar, np.sqrt(varfbar)
        
def plotting(X,Y,axis_x,axis_y,judul_tabel):
    fig = plt.figure(figsize=(10,5))
    plt.rcParams.update({'font.size': 15})
    plt.tick_params(colors='red', which='both')
    plt.title('{}'.format(judul_tabel), fontsize = 20, color = 'red')
    plt.xlabel('{}'.format(axis_x), fontsize = 20, color = 'red')
    plt.ylabel('{}'.format(axis_y), fontsize = 20, color = 'red')
    plt.plot(X,Y, 'o-', color = 'red')
    plt.grid()
    plt.show()
```

## Nomor 5

5. Jelaskan hasil perhitungan dengan kedua metode tersebut

Berdasarkan perhitungan mennggunakan metode analitik atau manual dengan library sympy diperoleh nilai hasil integral yaitu 0.499055289653755. Jika dibandingkan dengan 2 metode peng-integralan yaitu metode trapezoidal rule dan metode monte carlo maka akan menghasilkan nilai integral yang berbeda. Hasil integral dengan menggunakan metode trapezoidal rule dengan step h = 1 yaitu 0.587171985981535 dengan error trapezoidal sebesar 0.367926902547642. Namun, hasil integral menggunakan metode monte carlo yaitu 0.498739254681555 dengan error metode monte carlo sebesar 0.0011177780277117464. Dengan begitu, jika hasil integral dari kedua metode dibandingkan dengan metode analitik/manual memiliki perbedaan pada persen error yang dihasilkan. Perbandingan persen error trapezoidal dengan analitik/manual yaitu 17.657%. Sedangkan perbandingan persen error metode monte carlo dengan analitik/manual yaitu 0.063%. Oleh karena itu terlihat bahwa hasil integral metode monte carlo lebih mendekati dengan hasil analitik/manual sehingga persen error yang dihasilkan lebih kecil dari metode trapezoidal dan juga error yang yang dihasilkan lebih kecil dibandingkan dengan metode trapezoidal.
