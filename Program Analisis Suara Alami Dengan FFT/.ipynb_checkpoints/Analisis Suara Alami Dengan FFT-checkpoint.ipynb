## KOEFISIEN SENAR 1

```python
#Mencari berapa banyak koefisien dari audio tersebut
audio_senar_1,sr_senar_1 = librosa.load(os.path.join(base_dir,sinyal1))
print("Banyaknya koefisien sinyal awal : ",audio_senar_1.shape)

#Mencari koefisien transformasi fourier dari audio tersebut
audio_senar_1_fft = fft(audio_senar_1)
print("Banyaknya koefisien sinyal setelah transformasi :",audio_senar_1_fft.shape)

print("Banyaknya sampling rate dari audio :",sr_senar_1)
```

```python
#Mencari lamanya durasi audio
duration_1 = librosa.get_duration(audio_senar_1,sr_senar_1)
duration_1
```

```python
#Karena koefisien fourier sinyal yang di transformasi masih dalam bentuk bilangan kompleks, maka harus dibuat nilai absolutnya agar dalam bentuk bilangan
koefisien_senar_1 = np.abs(audio_senar_1_fft)
print("Koefisien fourier dari sinyal audio",koefisien_senar_1)
```

```python
data_senar_1 = pd.DataFrame((koefisien_senar_1))
data_senar_1.to_csv("FFT_suara_senar_1.csv")
data_senar_1=pd.read_csv('FFT_suara_senar_1.csv')
data_senar_1
```

```python
plt.figure(figsize=(10,5))
plt.plot(koefisien_senar_1)
plt.grid(True)
plt.xlim(100,1000)
```

## Energy Senar 1

```python
sigma_E_1 = (1/len(audio_senar_1)*sum(abs(np.square(audio_senar_1))))
Energy_senar_1 = 0
for i in (0,audio_senar_1):
    Energy_senar_1 += (1/len(audio_senar_1)*(abs(np.square(i))))
print("Nilai energy tiap iterasi adalah: ",Energy_senar_1)
print("Jumlah total energy dari semua iterasi adalah: ",sigma_E_1)
```

## Spectral Flux

## Spectral Flux Senar 1

```python
spectral_flux_senar_1 = 0
signal_senar_1 = koefisien_senar_1/sum(koefisien_senar_1)
eps_senar_1 = 2.2204e-16
for i in (0,koefisien_senar_1):
    signal_prev_senar_1 = (i-1)/sum((koefisien_senar_1-1)+eps_senar_1)
    spectral_flux_senar_1 += ((signal_senar_1-signal_prev_senar_1)**2)
print("Banyaknya spectral flux dari audio adalah: ",spectral_flux_senar_1.shape)
print("Total besar nilai spektral flux didapatkan: ",spectral_flux_senar_1.sum())
```

```python
spectral_flux_1 = pd.DataFrame((spectral_flux_senar_1))
spectral_flux_1.to_csv("fitur_SF_suara_senar_1.csv")
spectral_flux_1=pd.read_csv('fitur_SF_suara_senar_1.csv')
spectral_flux_1
```
