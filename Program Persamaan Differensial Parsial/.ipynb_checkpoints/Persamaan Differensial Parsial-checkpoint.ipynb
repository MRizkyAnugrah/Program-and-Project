3. Tuliskan syarat-syarat batas sesuai dengan gambar di atas (atas, bawah dan 
bagian terisolasi).

- Titik Terisolasi Atas dan Bawah

$$\frac{\partial^2T}{\partial y^2} = 0$$

- Titik Terisolasi Kanan

$$\frac{\partial^2T}{\partial x^2} = 0$$

- Titik Bagian Kiri 

$$\frac{\partial^2T}{\partial x^2} + \frac{\partial^2T}{\partial y^2} = 0$$
$$T =  120^{\circ}C - 0^{\circ}C = 120^{\circ}C$$

- Titik Bagian Kanan

$$\frac{\partial^2T}{\partial x^2} + \frac{\partial^2T}{\partial y^2} = 0$$
$$T = 0^{\circ}C$$

## Nomor 4

4. Tuliskan persamaan numerik untuk menghitung distribusi temperatur pada setiap 
pada plat (ada 36 titik) dengan menggunakan Metode Beda Hingga;

- Persamaan Laplace
$$\frac{\partial^2T}{\partial x^2} + \frac{\partial^2T}{\partial y^2} = 0$$

- Central Differences

$$\frac{\partial^2 T}{\partial x^2}=\frac{T_{i+1, j}-2 T_{i, j}+T_{i-1, j}}{\Delta x^2}$$
$$\frac{\partial^2 T}{\partial y^2}=\frac{T_{i, j+1}-2 T_{i, j}+T_{i, j-1}}{\Delta y^2}$$

- Subtitusikan nilai differensial orde 2 ke dalam rumus laplace

$$\frac{T_{i+1, j}-2 T_{i, j}+T_{i-1, j}}{\Delta x^2} + \frac{T_{i, j+1}-2 T_{i, j}+T_{i, j-1}}{\Delta y^2} =  0$$

- Pada kasus ini nilai $\Delta x = \Delta y$ maka

$$T_{i+1, j}+T_{i-1, j}+T_{i, j+1}+T_{i, j-1}-4 T_{i, j}=0$$
$$T_{i, j}=\frac{T_{i+1, j}+T_{i-1, j}+T_{i, j+1}+T_{i, j-1}}{4}$$

- Misalkan ialah titik i = 0, j = 0, maka:

$$T_{0,0}=\frac{T_{1,0}+T_{-1,0}+T_{0,1}+T_{0,-1}}{4}$$
$$T_{0,0}=\frac{T_{1,0}+T_{0,1}}{2}$$

```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

## Matriks

```python
Width = 7
Length = 9
matriks = []
for i in range(Width):
    list_kosong = []
    for j in range(Length):
        koordinat = [i,j]
        list_kosong.append(koordinat)
    matriks.append(list_kosong)
matriks
```

## Matriks Sesuai Soal

```python
for i in range(3):
    del matriks[i][5:9]
matriks
```

```python
matriks = np.zeros((Width,Length))
for i in range(Width):
    matriks[i][0] += 120 - 20*i
matriks
```

## Fungsi

```python
def central_differences(tabel,tabel_lama,i,j,Length,Width,Employ_overrelaxation=1.5):
    a = i + 1
    b = i - 1
    c = j + 1
    d = j - 1
    if a >= Width:
        a -= 2
    if b < 0:
        b += 2
    if c >= Length:
        c -= 2
    if d < 0 :
        d += 2
    list_kosong = np.zeros(1)
    list_kosong = round(1/4*(tabel[a][j] + tabel[b][j] + tabel[i][c] + tabel[i][d]),3)
    hasil = round(Employ_overrelaxation*list_kosong + (1-Employ_overrelaxation)*tabel_lama[i][j],3) 
    return round(hasil,4)
```

```python
def Error(tabel,tabel_lama,y,x):
    list_kosong = np.zeros(1)
    list_kosong = np.abs((tabel[y][x] - tabel_lama[y][x]) / tabel_lama[y][x])*100
    return list_kosong
```

## Menghitung T

```python
tabel = matriks.copy()
tabel_lama = matriks.copy()

for x in range(1,4):
    for y in range(Width):
        tabel[y][x] = round(central_differences(tabel,tabel_lama,y,x,Length,Width),3)
for x in range(4,5):
    for y in range(0,4):
        tabel[y][x] = round(central_differences(tabel,tabel_lama,y,x,5,Width),3)
for x in range(4,8):
    for y in range(4,Width):
        tabel[y][x] = round(central_differences(tabel,tabel_lama,y,x,Length,Width),3) 

for i in range(20):
    tabel_lama = tabel.copy()
    error = np.zeros((Width,Length))
    for x in range(1,4): 
        for y in range(Width):
            tabel[y][x] = round(central_differences(tabel,tabel_lama,y,x,Length,Width),3)
            error[y][x] = round(Error(tabel,tabel_lama,y,x),3)
    for x in range(4,5):
        for y in range(0,4):
            tabel[y][x] = round(central_differences(tabel,tabel_lama,y,x,5,Width),3)
            error[y][x] = round(Error(tabel,tabel_lama,y,x),3)
    for x in range(4,8):
        for y in range(4,Width):
            tabel[y][x] = round(central_differences(tabel,tabel_lama,y,x,Length,Width),3) 
            error[y][x] = round(Error(tabel,tabel_lama,y,x),3)

    sum_err = np.sum(error)
    rata = sum_err/37
    if rata < 1:
        Nilai_iter = i + 1
        break
```

```python
print('Soal \n', matriks)
print('Nilai T setiap titik \n', tabel)
print('Nilai Error \n', error)
```

```python
fig = plt.figure(figsize=(10,10))
fig.patch.set_facecolor('xkcd:white')
plt.subplot(211)
sns.heatmap(data = matriks, annot=True, fmt='g')
plt.title('Soal')
plt.subplot(212)
sns.heatmap(data = tabel, annot=True, fmt='g')
plt.title('Steady State')
plt.show()
```
